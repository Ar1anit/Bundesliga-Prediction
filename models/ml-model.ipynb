```python
import pandas as pd
```

Wir lesen hier unsere CSV-Datei ein, die die Daten der letzten Bundesliga Saisons (ab 19/20) hat

```python
matches = pd.read_csv('../datasets/merge.csv')
```

Mit matches.head() lassen wir uns die ersten 5 Zeilen der Tabelle anzeigen

```python
matches.head()
```

matches.shape zeigt uns die Länge der x- und y-Achse. Wir haben 105 Features (Spalten) die wir für unsere Prediction theoretisch verwenden können.

Wir schauen uns an wie häufig die Heimmanschaften in der Tabelle auftauchen.

```python
matches["HomeTeam"].value_counts()
```

Wir lassen uns die Heimspiele von Bayern anzeigen

```python
matches[matches["HomeTeam"] == "Bayern Munich"]
```

Mit .dtypes können wir uns die Datentypen der einzelnen Spalten anzeigen lassen. Das ist wichtig, weil manche Parameter als Objects gespeichert sind, sie wir aber als Zahl benötigen, da ein ML-Model wenig mit Objekten anfangen kann.

```python
matches.dtypes
```

Die Spalte Div ist unnötig, weil sie nur anzeigt um welche Liga es geht. Mit dem Drop dieser Spalte sparen wir uns Zeit beim iterieren der Datei.

```python
matches.drop(["Div"], axis="columns", inplace=True)
```

```python
matches.head()
```

Hier wandeln wir die Einträge in der Spalte "Datum" in ein passendes Format um, das vom Model gelesen werden kann.

```python
matches["Date"] = pd.to_datetime(matches["Date"])
```

```python
matches.head()
```

Wir erstellen eine neue Spalte "FTR_code" in der die Spalte FTR (Full Time Result) als Zahl dargestellt wird, für jedes Spiel.

```python
matches["FTR_code"] = matches["FTR"].astype("category").cat.codes
```

```python
matches.head()
```

```python
matches["HTR_codes"] = matches["HTR"].astype("category").cat.codes
```

```python
matches.head()
```

Genauso bekommt jeder Gegner einen Code:

```python
matches["opp_codes"] = matches["AwayTeam"].astype("category").cat.codes
```

```python
matches.head()
```

```python
matches["team_codes"] = matches["HomeTeam"].astype("category").cat.codes
```

```python
matches.head()
```

Die Zeit muss auch noch in eine Zahl umgewandelt werden, damit wir damit arbeiten können.

```python
matches["hour"] = matches["Time"].str.replace(":.+", "", regex=True).astype("int")
```

```python
matches.head()
```

Wir fügen noch einen day_code hinzu der die Wochentage als Zahlen kodiert.

```python
matches["day_code"] = matches["Date"].dt.day_of_week
```

```python
matches.head()
```

Wir erstellen eine Spalte "target" mit den Werten 0 und 1. 0 bedeutet, dass die Heimmannschaft verloren hat oder unentschieden, 1 bedeutet, dass die Heimmannschaft gewonnen hat.

```python
matches["target"] = (matches["FTR"] == "H").astype("int")
```

```python
matches.head()
```

Wir importieren den RandomForestClassifier aus dem sklearn Paket.

```python
from sklearn.ensemble import RandomForestClassifier
```

Wir erstellen eine Instanz des Random Forest Klassifikators. n_estimators gibt an, wie viele Bäume wir erstellen wollen. max_depth gibt an, wie tief die Bäume sein dürfen. random_state ist ein Seed, damit wir die Ergebnisse reproduzieren können.

```python
rf = RandomForestClassifier(n_estimators=100, min_samples_split=10, random_state=0)
```

Wir defnieren unseren Trainings- und Testdatensatz. Unser Trainingsdatensatz sind die Spiele vor dem 27.05.2023. Unsere Testdatensatz sind die Spiele der bisherigen Saison 23/24.

```python
train = matches[matches["Date"] < "27-05-2023"]
```

```python
test = matches[matches["Date"] > "27-05-2023"]
```

```python
test.dropna(inplace=True)
```

```python
train.dropna(inplace=True)
```

Hier definieren wir unsere predictors. Sie sind die Features, die wir für die Vorhersage verwenden wollen. 

```python
predictors = [
    "team_codes", 
    "opp_codes", 
    "hour", 
    "day_code",
    "B365H", "B365D", "B365A", 
    "BWH", "BWD", "BWA", 
    "IWH", "IWD", "IWA", 
    "PSH", "PSD", "PSA", 
    "WHH", "WHD", "WHA", 
    "VCH", "VCD", "VCA", 
    "MaxH", "MaxD", "MaxA"
]
```

Mit rf.fit trainieren wir das Modell mit den Trainingsdaten.

```python
rf.fit(train[predictors], train["target"])
```

preds ist hier die Variable, die die Vorhersagen des Modells enthält: 

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

Die Variable acc ist die Genauigkeit des Modells. Sie gibt an, wie oft das Modell richtig vorhersagt, ob die Heimmannschaft gewinnt oder nicht:

```python
acc = accuracy_score(test["target"], preds)
```

```python
acc
```

```python
combined = pd.DataFrame(dict(actual=test["target"], prediction=preds))
```

```python
pd.crosstab(index=combined["actual"], columns=combined["prediction"])
```

```python
from sklearn.metrics import precision_score
```

```python
precision_score(test["target"], preds)
```

```python
grouped_matches = matches.groupby("HomeTeam")
```

```python
group = grouped_matches.get_group("Bayern Munich")
```

```python
group
```

```python
def rolling_average(group, cols, new_cols):
    group = group.sort_values("Date")
    rolling_stats = group[cols].rolling(5, closed="left").mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
matches.columns
```

```python
cols = ["FTHG", "FTAG", "HTHG", "HTAG", "HS", "AS", "HST", "AST", "HC", "AC", "HF", "AF", "HY", "AY", "HR", "AR"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_average(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("HomeTeam").apply(lambda x: rolling_average(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel("HomeTeam")
```

```python
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
def make_prediction(data, predictors):
    train = data[data["Date"] < "27-05-2023"]
    test = data[data["Date"] > "27-05-2023"]
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], prediction=preds))
    precision = precision_score(test["target"], preds)
    return combined, precision
```

```python
matches_rolling.dropna(inplace=True)
```

```python
combined, precision = make_prediction(matches_rolling, predictors + new_cols)
```
